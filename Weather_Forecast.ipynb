#Assigning,filtering,appointing the data as per need:

```python
df = df.drop('play', axis=1, errors='ignore')
df
```

```python
df.rename(columns={'outlook':'Weather'},inplace=True)
df.rename(columns={'temperature':'temperature(Celcius)'},inplace=True)
df.rename(columns={'humidity':'humidity(in %)'},inplace=True)
df
```

##For weather:

Hot: Around 30°C to 40°C or higher.
(Example: A sunny summer day in a desert region.)

Mild: Around 15°C to 25°C.
(Example: A pleasant spring or autumn day, neither too warm nor too cold.)

Cool: Around 10°C to 15°C.
(Example: A crisp autumn morning or early spring day where you might need a light jacket.)

```python
def oye(a):
  if(a=='hot'):
    return 35
  if(a=='mild'):
    return 25
  if(a=='cool'):
    return 15
df['temperature(Celcius)']=df['temperature(Celcius)'].apply(oye)
df
```

##For Humidity:

High humidity:

Above 70% relative humidity is generally considered high.
(Example: A humid, sticky day that feels uncomfortable and muggy.)

For weather (outdoor air):

Normal humidity is around 30% to 60% relative humidity.
(This range is comfortable and doesn't feel too dry or too humid.)

```python
def hum(c):
  if(c=='high'):
    return 70
  if(c=='normal'):
    return 45
df['humidity(in %)']=df['humidity(in %)'].apply(hum)
df
```
